# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]

urban_ride_count = urban_cities_df.groupby(["city"]).count()["ride_id"]
suburban_ride_count = suburban_cities_df.groupby(["city"]).count()["ride_id"]
rural_ride_count = rural_cities_df.groupby(["city"]).count()["ride_id"]

urban_ride_total = urban_ride_count.sum()
suburban_ride_total = suburban_ride_count.sum()
rural_ride_total = rural_ride_count.sum()

total_rides = pd.DataFrame({'type':['Rural', 'Suburban', 'Urban'],'Total Rides':[rural_ride_total, suburban_ride_total, urban_ride_total]})
total_rides = total_rides.set_index('type')

total_rides
```

```python
# 2. Get the total drivers for each city type
sum_drivers_by_type = city_data_df.groupby(["type"]).sum()["driver_count"]

total_drivers = pd.DataFrame({'type':['Rural', 'Suburban', 'Urban'],'Total Drivers':[sum_drivers_by_type[0],sum_drivers_by_type[1],sum_drivers_by_type[2]]})
total_drivers = total_drivers.set_index('type')

total_drivers
```

```python
#  3. Get the total amount of fares for each city type
sum_fares_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]

total_fares = pd.DataFrame({'type':['Rural', 'Suburban', 'Urban'],'Total Fares':[sum_fares_by_type[0],sum_fares_by_type[1],sum_fares_by_type[2]]})
total_fares = total_fares.set_index('type')

total_fares
```

```python
#  4. Get the average fare per ride for each city type. 
avg_rural_fares = sum_fares_by_type[0] / rural_ride_total
avg_suburban_fares = sum_fares_by_type[1] / suburban_ride_total
avg_urban_fares = sum_fares_by_type[2] / urban_ride_total

avg_fares = pd.DataFrame({'type':['Rural', 'Suburban', 'Urban'],'Average Fare per Ride':[avg_rural_fares,avg_suburban_fares,avg_urban_fares]})
avg_fares = avg_fares.set_index('type')

avg_fares
```

```python
# 5. Get the average fare per driver for each city type. 
driver_rural_fares = sum_fares_by_type[0] / sum_drivers_by_type[0]
driver_suburban_fares = sum_fares_by_type[1] / sum_drivers_by_type[1]
driver_urban_fares = sum_fares_by_type[2] / sum_drivers_by_type[2]

driver_avg_fares = pd.DataFrame({'type':['Rural', 'Suburban', 'Urban'],'Average Fare per Driver':[driver_rural_fares,driver_suburban_fares,driver_urban_fares]})
driver_avg_fares = driver_avg_fares.set_index('type')

driver_avg_fares
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({'type':['Rural', 'Suburban', 'Urban'],'Total Rides':[rural_ride_total, suburban_ride_total, urban_ride_total],'Total Drivers':[sum_drivers_by_type[0],sum_drivers_by_type[1],sum_drivers_by_type[2]],'Total Fares':[sum_fares_by_type[0],sum_fares_by_type[1],sum_fares_by_type[2]],'Average Fare per Ride':[avg_rural_fares,avg_suburban_fares,avg_urban_fares],'Average Fare per Driver':[driver_rural_fares,driver_suburban_fares,driver_urban_fares]})
pyber_summary_df = pyber_summary_df.set_index('type')
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].astype(float)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].astype(float)
#pyber_summary_df["Total Rides"][2] = int(pyber_summary_df["Total Rides"][2])
#pyber_summary_df["Total Drivers"][0] = pyber_summary_df["Total Drivers"][0].astype(float)
#pyber_summary_df["Total Drivers"][1] = pyber_summary_df["Total Drivers"][1].astype(float)
#pyber_summary_df["Total Drivers"][2] = pyber_summary_df["Total Drivers"][2].astype(int)
#pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].astype(float)
#pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].astype(float)
#pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].astype(float)

pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
type_date = pyber_data_df.groupby(["type","date"]).sum()["fare"]
type_date
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
type_date = type_date.reset_index()
type_date.head()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
type_date_pivot = type_date.pivot(index="date", columns="type", values="fare")
type_date_pivot.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
early2019_data = type_date_pivot.loc['2019-01-01':'2019-04-29']
early2019_data.head(10)
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
early2019_data.index = pd.to_datetime(early2019_data.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
early2019_data.info
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
early2019_data_re = early2019_data.resample('W').sum()
early2019_data_re.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
weekly_fares = early2019_data_re.plot(figsize = (20,6))
# Add a title 
weekly_fares.set_title("Total Fare by City Type")
#  Add a x- and y-axis label.
weekly_fares.set_xlabel("Month")
weekly_fares.set_ylabel("Fare ($USD)")

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

# Save figure
plt.savefig("analysis/PyBer_fare_summary.png")

plt.show()
```
